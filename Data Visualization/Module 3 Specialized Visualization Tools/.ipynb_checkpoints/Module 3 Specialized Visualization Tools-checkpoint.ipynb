```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel('Canada.xlsx', sheet_name='Canada by Citizenship', skiprows=range(20), skipfooter=2);
```

```python
data.head()
```

```python
data.drop(['Type','Coverage','AREA','REG','DEV'], axis=1, inplace=True)
# let's rename the columns so that they make sense
data.rename(columns={'OdName':'Country', 'AreaName':'Continent','RegName':'Region'}, inplace=True)
```

```python
data.rename(columns={'OdName':'Country', 'AreaName':'Continent','RegName':'Region'}, inplace=True)
data.columns = list(map(str, data.columns))
data.set_index('Country', inplace=True)
data['Total'] = data.sum(axis=1)
data.head()
```

```python
years = list(map(str, range(1998, 2014)))
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')
```

```python
df_continents = data.groupby('Continent', axis=0).sum()
df_continents
```

```python
df_continents['Total'].plot(kind='pie',
                           figsize=(10, 5),
                           autopct='%1.1f%%',
                           startangle=90,
                           shadow=True)
plt.title('Immigration to Canada by Continent [1980 - 2013]')
plt.axis('equal') # Sets the pie chart to look like a circle.
plt.show()
```

```python
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1] # ratio for each continent with which to offset each wedge.
# df_continents
df_continents['Total'].plot(kind='pie',
                   figsize=(10, 5),
                   shadow=True,
                   autopct='%1.1f%%',
                   startangle=45,
                   labels=None,
                   pctdistance=1.13,
                   colors=colors_list,
                   explode=explode_list,
                  )
plt.title('Immigration to Canada by Continent [1980 - 2013]', y=1.12) 
plt.axis('equal') 
plt.legend(labels=df_continents.index, loc='upper left')
plt.show()
```

```python
explode_list = [0.0, 0, 0, 0.1, 0.1, 0.2] # ratio for each continent with which to offset each wedge.
df_continents['2013'].plot(kind='pie',
                   figsize=(15, 6),
                   shadow=True,
                   autopct='%1.1f%%',
                   startangle=90,
                   labels=None,
                   pctdistance=1.13,
                   explode=explode_list,
                  )
plt.title('Immigration to Canada by Continent in 2013', y=1.12) 
plt.axis('equal') 
plt.legend(labels=df_continents.index, loc='upper left') 
plt.show()
```

```python
df_top15 = data.sort_values(['Total'], ascending=False, axis=0).head(15)
```

```python
new_df.plot(kind='box', figsize=(15, 6), vert=False)
plt.title('Immigration from top 15 countries for decades 80s, 90s and 2000s')
plt.show()
```

```python
df_tot = pd.DataFrame(data[years].sum(axis=0))

df_tot.index = map(int, df_tot.index)
df_tot.reset_index(inplace = True)

df_tot.columns = ['years', 'Total']
df_tot.head()
```

```python
df_tot.plot(kind='scatter', x='years', y='Total', figsize=(10, 6), color='darkblue')
plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.show()
```

```python
x = df_tot['years']
y = df_tot['Total']
fit = np.polyfit(x, y, deg=1)
fit
```

```python
df_tot.plot(kind='scatter', x='years', y='Total', figsize=(10, 6), color='darkblue')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

plt.plot(x, fit[0] * x + fit[1], color='red')
plt.annotate('y={0:.0f} x + {1:.0f}'.format(fit[0],fit[1]), xy=(2000, 150000))
plt.show()
```

```python
# print out the line of best fit
'No. Immigrants = {0:.0f} * Year + {1:.0f}'.format(fit[0], fit[1])
```

```python
df_tot3 = pd.DataFrame(data.loc[['Denmark','Norway','Sweden'], years].sum(axis=0))
df_tot3.index = map(int, df_tot3.index)
df_tot3.reset_index(inplace=True)
df_tot3.columns = ['Years','Total']
df_tot3['Years'] = df_tot3['Years'].astype(int)
df_tot3.head()
```

```python
df_tot3.plot(kind='scatter', x='Years', y='Total', figsize=(10, 5), color='blue')

plt.title('Total Immigration of Denmark, Norway & Sweden to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

plt.show()
```

```python
data_t = data[years].transpose()
data_t.index = map(int, data_t.index)

data_t.index.name = 'Years'

data_t.reset_index(inplace=True)

data_t.head()
```

```python
norm_brazil = (data_t['Brazil'] - data_t['Brazil'].min()) / (data_t['Brazil'].max() - data_t['Brazil'].min())

norm_argentina = (data_t['Argentina'] - data_t['Argentina'].min()) / (data_t['Argentina'].max() - data_t['Argentina'].min())
```

```python
ax0 = data_t.plot(kind='scatter',
                  figsize=(10, 5),
                  x='Years',
                  y='Brazil',
                  alpha=0.5,
                  color='green',
                  s=norm_brazil * 2000 + 10,
                  xlim=(1975, 2015)
                 )

ax1 = data_t.plot(kind='scatter',
                  figsize=(10, 5),
                  x='Years',
                  y='Argentina',
                  alpha=0.5,
                  color='blue',
                  s=norm_argentina * 2000 + 10,
                  ax=ax0
                 )

ax0.set_title('Canada immigration from Brazil and Argentina from 1980 - 2013')
ax0.set_ylabel('Countries')
ax0.legend(['Brazil','Argentina'], loc='upper left', fontsize='x-large')
```
