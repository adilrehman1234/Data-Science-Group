```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
print(plt.style.available)

mpl.style.use(['ggplot'])
```

```python
years = data.iloc[0:, 2:-1]
years
```

```python
Region = years.loc['Eastern Africa']
Region.plot()
```

```python
Region.index = Region.index.map(int)
Region.plot(kind='line')
plt.title('Imigration from Eastern Africa')
plt.xlabel('Years')
plt.ylabel('Number of imigrations')
plt.show()
```

```python
Region.plot(kind='line')
plt.title('Imigration from Eastern Africa')
plt.xlabel('Years')
plt.ylabel('Number of imigrations')
plt.text(1980, 8000, 'Imigration increases')
plt.show()
```

```python
MiddleAfrica = years.loc['Middle Africa']
MiddleAfrica.index = MiddleAfrica.index.map(int)
MiddleAfrica.head()
```

```python
MiddleAfrica.plot(kind='line')
plt.title('Imigration from Middle Africa')
plt.xlabel('Years')
plt.ylabel('Number of imigrations')
plt.text(1985, 4050, 'Imigration continously increasing')
plt.show()
```

```python
Asia = years.loc['Eastern Asia']
Asia.index = Asia.index.map(int)

Asia.plot(kind='line')
plt.title('Imigration from Eastern Africa')
plt.xlabel('Years')
plt.ylabel('Number of imigrations')
plt.text(1980, 40500, 'Imigration flow of Eastern Asia')
plt.show()
```

```python
df_CI = years.iloc[1:6]
df_CI
```
